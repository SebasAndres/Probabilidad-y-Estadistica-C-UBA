## Guia 6 ~ Estadistica Descriptiva

### Mediana ($\widetilde X$):

> $Med(X) = x^{(k+1)}$ si $n=2k+1$<br>
> $Med(X) = \frac{x^{(k)} + x^{(k+1)}}{2}$ si $n = 2k$

### Media Muestral ($\bar X$):

> $\bar X = \frac{1}{n} \sum_{i=1}^n X_i$

### Varianza Muestral ($S^2$):

> $S^2 = \frac{1}{n-1} \sum (X_i - \bar X)^2$

### Distancia Absoluta de la Mediana (MAD):
> 1. Tomo la mediana de la muestra $Med(X) = \widetilde X = Med(x^{(1)}, ..., x^{(N)})$
> 2. Armo una nueva muestra $Y$ aplicando $|X_i - \widetilde X| \rightarrow Y = (|x^{(1)} - \widetilde x|, ..., |x^{(N)} - \widetilde x|)$
> 3. Tomo la mediana de esta nueva muestra $Med(Y) = \widetilde Y$
> 4. Concluyo $MAD(X) = Med(Y) = \widetilde Y$

### Media $\alpha$-podada:

> Es el resultado de calcular la media al conjunto de datos proviniente de eliminar $\alpha * n$ datos de ambos extemos, de la muestra de $X_i$.

### Distribucion empirica:

> Es ver #{casos favorables} / #{casos totales}

## Guia 7 ~ Estimadores

### Maxima Verosimilitud:

1. Armo la funcion de verosimilitud $L(X, \theta) = \prod_{i=1}^n f_{x, \theta}(x_i)$
2. Busco el maximo de $L$ respecto a $\theta$, llamemoslo $K$, pudiendo usar Log-Likelihood: $l(X, \theta) = Ln(L(X, \theta))$
3. $\hat \theta_{MV} = K$

> Tips:<br>
a. Para el maximo mirar $\frac{\partial{l}}{\theta} = 0 \land \frac{\partial^2{l}}{\theta} < 0$<br>
b. $\prod e^{\frac{1}{2}(\frac{X_i - \mu}{\sigma})^2} = e^{\frac{1}{2\sigma^2}\sum (X_i -\mu)^2}$<br>
c. Cuando $f_x(x)$ tiene una indicadora para un intervalo $I$, y este depende del parametro, separar a $L(X)$ en el caso que todos los $X_i \in I$ (cuando $L=g(X, \theta)$) y el que no ($L=0$).
Luego:
> * Fijarse que debe cumplir $\theta$ en ese intervalo,
    ej. si $I=[0,7\theta] \rightarrow 7\theta \geq Max(X_1, ..., X_n) \rightarrow \theta \geq \frac{Max(X_1, ..., X_n)}{7}$ 
> * Fijarse si $L(X, \theta) \rightarrow 0$ para $\theta \rightarrow +inf$, si se da ese caso tomar el $\theta$ minimo que cumpla el intervalo. 
    En nuestro ejemplo, $\theta_{MV} = \frac{Max(X_1, ..., X_n)}{7}$


### Momentos:
1. Armo la esperanza de la variable aleatoria.
2. La comparo con la media muestral.
3. Despejo el parametro de interes (que aparece en la esperanza calculada)

### Sesgo 
> $b(\hat \theta_n) = E(\hat \theta_n) - \theta$

### Consistencia:
> Es consistente $\leftrightarrow \hat \theta_n \rightarrow \theta$ en probabilidad <br>

> Tip: Si $\hat \theta_n = - \frac{1}{n} \sum ln(x_i)$, con ... <br>
> * $Y_i = ln(x_i)$<br>
> * $E(ln(x_i)) = -\theta$ <br>
> $\hat \theta_n = - \frac{1}{n} \sum Y_i = -\hat Y_n \rightarrow -E(Y)=-(-\theta) = \theta$ <br>
> Luego $\hat \theta_n$ es consistente. 

> Si no puedo probarlo de esta forma, uso el resultado...<br>
> $\hat \theta $ es consistente $\leftrightarrow$, cuando $n \rightarrow inf$<br>
> * $E(\hat \theta) \rightarrow  \theta$ <br>
> * $V(\hat \theta) \rightarrow  0$ <br>

### Error Cuadratico Medio
> $ECM(\hat \theta) = E([\hat \theta_n - \theta]^2) = V(\hat \theta_n) + b(\hat \theta_n)^2$

### Tips en IPP
> Si la funcion tiene la forma $e^{-t(x)}$, hacer un cambio de variable con $t(x)$ dejando afuera el menos<br>
> $dv = e^{-t}$ y $u=t$<br>
> $u=ln(x)$ y $dv=x^{(\frac{1}{\theta} -1)}$<br>
> Si queda una indeterminacion, tratar de escribirla como division y usar L'Hopital

## Guia 8 ~ Intervalos de Confianza

### Intervalo de Confianza, la idea...
> 1. Armo el estadistico $T$ ~ $F$
> 2. Se que $P(-T_{\alpha/2} <= T <= T_{\alpha/2}) = 1 - \alpha$
> 3. Despejo el parametro de interes en funcion a la muestra y los $T_\alpha$
> 4. Reescribo el intervalo

### Estadisticos a saber.

Sea $X$ ~ $N(\mu, \sigma^2)$:

> IC para $\mu$ con $\sigma$ conocida: <br>
> * $\frac{\bar X - \mu}{\sigma}\sqrt n$ ~ $N(0,1)$

> IC para $\mu$ con $\sigma$ desconocida: <br>
> * $\frac{\bar X - \mu}{S}\sqrt n$ ~ $t_{n-1}$

> IC para $\sigma$ con $\mu$ conocida: <br>
> * $\sum (\frac{X_i - \mu_0}{\sigma})^2 $ ~ $\chi_n^2$

> IC para $\sigma$ con $\mu$ desconocida: <br>
> * $\frac{(n-1)S^2}{\sigma^2}$ ~ $\chi_{n-1}^2$

Sea $X$ ~ $Bi(n,p)$:
> IC (aproximado) para una proporcion $p$:
> * $\frac{p - \bar X}{\sqrt{\bar X(1-\bar X)}}\sqrt n$ ~ $N(0,1)$

Sea $X$ ~ $F$:
> * $\frac{\bar X - \mu}{\theta}\sqrt n$ ~ $N(0,1)$, aproximadamente <br>
> * $\frac{\bar X - \mu}{S}\sqrt n$ ~ $N(0,1)$, aproximadamente

### Tips:
> <u>Ejercicio:</u> Hallar el $n$ minimo para que el $IC_{1-\alpha}$ del parametro $p$, de nivel aproximado $1-\alpha$, tenga longitud menor a  0.1.<br>
> <u>Solucion:</u>
> * Sea $L$ la longitud, por el intervalo de confianza (centrado en $\bar X$), <br>
> * $L = 2 * Z_{\alpha/2}  \sqrt{\frac{\bar X_n (1 - \bar X_n)}{n}} \leq 0.1$
> * En particular, puedo acotar $\bar X_n (1 - \bar X_n) \leq 0.25$
> * $L = 2 * Z_{\alpha/2} \sqrt{\frac{0.25}{n}} \leq 0.1$
> * $\sqrt n \geq \frac{2 * Z_{\alpha/2} \sqrt {0.25}}{0.1}$
> * Concluyo, $n \geq (\frac{2 * Z_{\alpha/2} \sqrt {0.25}}{0.1})^2$

## Guia 9 ~ Test de hipotesis

Busco evidencia para rechazar o no hipotesis sobre el valor de algun parametro de una distribucion.

### 1. Armado del test 
> Hipotesis nula ($H_0$): La quiero rechazar. (Ej. $H_0: \mu = 4$). <br>
> Hipotesis del investigador ($H_1$): Algo de lo que estoy buscando evidencia. (Ej. $H_1: \mu < 4$).

Observacion: la $H_0$ debe tener un igual.

### 2. Eleccion del estadistico:
> Elijo un estadistico que no dependa del parametro que busco<br>
> (Ej. $T = \frac{\bar X - \mu}{\sigma}\sqrt n$ ~ $N(0,1)$)

### 3. Armo la zona de rechazo (RR):
> (Ej: $RR$ : {$T_{obs} | T_{obs} > T_\alpha$})

Obseracion: Si $H_1: \mu \neq \mu_0 \rightarrow RR$ : {$T_{obs}: |T_{obs}| > T_{\alpha / 2}$}

### 4. Calculo $T_{obs}$ y me fijo si $T_{obs} \in RR$.
> Si $T_{obs} \in RR \rightarrow $ rechazo $H_0$  
> Si $T_{obs} \notin RR \rightarrow $ no rechazo $H_0$, me falta evidencia  

### P-valor:
El p-valor es la probabilidad de que el estadistico observado tenga un valor mas extremo de lo observado. <br>
Si p-valor $\rightarrow 0$, entonces tengo mas evidencia para rechazar $H_0$ <br>
Si p-valor $\leq T_{obs}$, rechazo $H_0$

Casos:
* a. $H_1: \mu > \mu_0 \rightarrow p = P(T>T_{obs})$
* b. $H_1: \mu < \mu_0 \rightarrow p = P(T<T_{obs})$
* c. $H_1: \mu \neq \mu_0 \rightarrow p = P(|T|>T_{obs}) = P(T > T_{obs}) + P(T < -T_{obs})$ 

### Errores:
> Tipo I: Rechazo $H_0$ con $H_0$ verdadera.<br>
> Tipo II: No rechazo $H_0$ con $H_0$ falsa.

### Funcion de potencia:
> $\pi(\mu_0) : $ Probabilidad de rechazar $H_0$ dado $\mu = \mu_0$. <br>
> Puede pensarse como una funcion partida:<br>
> * $\theta \in H_0 \rightarrow \pi(\theta) = \alpha(\theta)$<br>
> * $\theta \in H_1 \rightarrow \pi(\theta) = 1 - \beta(\theta)$

Observacion: Podria ser cualquier otro parametro, no solo $\mu$.

### Tips:

#### Tip 1
> Si me pide la probabilidad de rechazar $H_0$, con $\mu = \mu_0$, me fijo... <br>
> $\pi(\mu) = 1 - \beta$<br>
> $\beta = P(Error II) = P_{\mu=\mu_0}(T_{obs} \notin RR)$...

#### Tip 2: 
> Si me da un p-valor ($p_0 = 0.10$) para un test de la forma: <br>
> $H_0: \mu = \mu_0$ vs $H_1: \mu \neq \mu_0$ <br>
> Y me pregunta posibles valores de $\bar X_{obs}$, miro dos casos. <br>
> Como $p$ es la probabilidad de tener un valor mas extremo en el estadistico, puede ser
> * Caso 1:
>   * $p = P(T > T_{obs}) = p_0 = 0.10$
>   * y despejo de $\bar X_n$ a la muestra.
> * Caso 2:
>   * $p = P(T < T_{obs}) = p_0 = 0.10$
>   * y despejo de $\bar X_n$ a la muestra.
